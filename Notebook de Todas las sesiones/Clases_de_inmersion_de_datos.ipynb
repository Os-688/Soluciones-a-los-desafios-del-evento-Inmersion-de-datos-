```python
barrio_area = inmuebles.loc[:, ["Barrio", "Area"]]
```

```python
agrupacion_barrio_V = barrio_area.groupby("Barrio")
agrupacion_barrio_V.value_counts()
```

```python
promedio_area = agrupacion_barrio_V.mean(numeric_only = True)
promedio_area.plot.bar()
```
