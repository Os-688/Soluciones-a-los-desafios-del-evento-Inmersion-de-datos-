**Creación de Rangos:**

Min 60 max 16000 media 602  75 680

rangos:
rango 1 de 60 a 300 -- medio bajo
rango 2 de 301 a 602 -- medio
rango 3 de 603 a 900 -- alto
rango 4 mayor de 900 -- desorbitado

```python
inmuebles["Division_precio"] = inmuebles["Precio_Millon"].apply(lambda x:"Medio bajo" if x >= 60 and x<=300
                                                                            else ("Medio" if (x >= 301 and x <= 602)
                                                                            else ("Alto" if (x >=603 and x <= 900)
                                                                            else "Desorbitado")))

# Uso la lambda if para realizar los rangos y agregarlos a una variable
```

```python
inmuebles["Valor m2"] = (inmuebles.Precio / inmuebles.Area).round(2)
# Creo una variable del valor m2 sin millon
```

```python
inmuebles.sample(10)
```

```python
tres = (inmuebles["Tipo"] == "Apartamento") | (inmuebles["Tipo"] == "Casa") | (inmuebles["Tipo"] == "Finca")
#Creo el booleano para los tres tipos de inmuebles y lo aplico
```

```python
sum((inmuebles["Tipo"] == "Bodega"))
```

```python
tipos_inmuebles = inmuebles[tres]
tipos_inmuebles
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# La grafica demuestra la frecuencia de los rangos de precio por tipo. Valor que da una comparacion entre tipos y rangos.


sns.color_palette("mako", as_cmap=True)
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightgray'})

sns.color_palette("mako")

plt.figure(figsize=(10,6))
grafica = sns.histplot(data=tipos_inmuebles, x='Tipo', kde=True, hue='Division_precio', multiple="stack")

sns.set(font_scale=1.4)# Cambia el tamaño de todos los elementos por cambiar la fuente de tamaño

#plt.xlim((50,1000))
plt.ylim(0,8000)

plt.xlabel('Tres tipos', fontsize=14);
plt.ylabel('Frecuencia', fontsize=14);
plt.title("Comparacion de frecuencia de precio de Casa, Apartamento, Finca.", fontsize=16)

plt.savefig('/content/drive/MyDrive/Evento_inmersion_de_datos/comparacion de frecuencias en precios',format='png')
plt.show()
```

### Desafio 1: <sup>Probar con otros modelos de ML:<sup>

***Modelos regresivos y clasificadores:***

Los modelos regresivos trabajan con numeros flotantes y los modelos clasificadores solo con enteros. Esto hace que los modelos de Support Vector Machine y Linear regresion funcionen con datos distintos. Ya que el primero es un clasificador y predice datos enteros. Y el segundo sea un modelo regresivo por lo tanto pueda predecir datos no tan exactos como numeros flotantes.

Entonces para el siguiente modelo de SVC no se puede usar las variable Y que se uso en el entrenamiento porque la variable es un double. Esto seria igual para un Arbol de decisiones.

*Datos continuous y multicast:*

Los datos continuous son los que se usan en modelos regresivos y los multicast en clasificatorios.

**Support Vector Machine**

```python
datos_ml.columns
```

**Con datos flotantes Resumen:**

En este caso use un metodo de sklearn para realizar un cast de un dato flotante a un dato entero sin tantas perdidas como un casta natural de python. Esto lo hice usando una biblioteca de sklearn para transformar estos datos que se llaman continuous a un dato multiclass.

De esta forma se ve que un modelo que trabaja con datos multiclass puede trabajar con datos con mucha distincion como los continuous (Un ejemplo son los flotantes), aunque no se hasta que punto es correcto porque los modelos con datos continuous estan diseñados para este tipo de datos con pocas agrupaciones y los modelos clasificatorios usan agrupaciones para generar predicciones. Creo que tendra sus fallos y carencias. Pude observarlo porque probe con la variable Division_precio y el porcentaje de precision fue el muy distinto con los mismos datos de prueba. Aunque probe otra variable porque queria obtener las variables adecuadas para predecir pero no lo logre. Porque la funcion de correlaciones no plantea a division-precio en el esquema, asi que tuve que cambiar de variable.

**Conclusion:**

En definitiva los resultados usando un modelo REGRESIVO para los datos continuous son mejores que los que se pueden obtener con un modelo que funciona por agrupaciones como SVC. Así que si los modelos se desempeñan mejor con los datos por los cuales fueron diseñados.

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn import utils # Esta bibibioteca sirve para ver si es un dato multiclass o continuous


# x = datos_ml[['COD_UPZ_GRUPO','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']]

# Antes eran estas variables pero cambie a estas por que tenian mejor desempeño

x = datos_ml[['Area','Banos',"Valor_m2_Millon","Valor_m2_Barrio","SALARIO_ANUAL_MI","RUIDO"]]

y = datos_ml["Precio_Millon"]

lab_enc = preprocessing.LabelEncoder()
# creando el objeto que casteara a numero mas simple pero sin perder tanta informacion

# cast de los datos de clasificacion
encoded = lab_enc.fit_transform(y)

#Estableciendo semilla
SEED = 76
np.random.seed(SEED)

# Creando la carga de prueba y entrenamiento, estas son variables que se pueden usar. Ya estan preparadas
raw_x_train, raw_x_test, y_train, y_test = train_test_split(x,encoded,test_size = 0.25)

# Creando escalador para optimizar los datos para el modelo

scaler = StandardScaler()
scaler.fit(raw_x_train)

# Transdormando los datos con el escalador ya preparado
x_train = scaler.transform(raw_x_train)
x_test = scaler.transform(raw_x_test)

# Creacion del modelo, entrenamiento y predicciones
model = SVC()
model.fit(x_train, y_train)
previsiones = model.predict(x_test)


# Usando el metodo accuracy para obtener la tasa de acierto
tasa_de_acierto = accuracy_score(y_test, previsiones)
print(f"La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%")
```

```python
# Baseline: linea base para comparar la prediccion del modelo SVC y el tipo de datos de arriba con los de esta prediccion sencilla o estadistica.

from sklearn.dummy import DummyClassifier
dummy = DummyClassifier(strategy="most_frequent")
# Dummy con estrategia most frequent, la cual es mas efectiva que  estratified
dummy.fit(x_train,y_train)
exactitud = dummy.score(x_test, y_test)*100
print(f"La exactitud del clasificador Dummy stratified fue: {round(exactitud, 2)}%")
```

**Con datos Multiclass (sin datos flotantes) Resumen:**

Los datos con multiclass funcionan de manera mas sencilla en modelos clasificadores y tienen un mayor desempeño en alcanzar la optimizacion mas alta en este modelo SVC. Tambien puedo decir que para la prediccion de habitaciones a funcionado con mayor precision que Precio Millon e incluso que la variable Division_precio. Esto ultimo podria significar que hay una mejor optimizacion de variables para division_precio.

***Conclusion:***

Los datos superan a la linea base, aunque esta es un dummy classifier por lo que sus resultados se ven un poco bajos en comparacion con el modelo SVC, aunque sirve para comparar en un principio, puesto que cuando probemos otro modelo u ajuste de variables o parametros se podra comparar con el dummy classifier, el modelo SVC u Otros ajustes en el mismo modelo o con otros modelos los que se vayan probando.
